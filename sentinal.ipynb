```python
coord_cols = ['X_COORD_CD', 'Y_COORD_CD', 'Latitude']
for col in coord_cols:
    # Remove any stray text like "approx" before converting
    df[col] = df[col].str.extract(r'([-+]?\d*\.?\d+)', expand=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# NYC valid latitude range is roughly 40.4 to 40.9
# Null Island: lat = 0.0, or completely out of NYC range
df.loc[df['Latitude'] == 0.0, ['X_COORD_CD', 'Y_COORD_CD', 'Latitude']] = np.nan
df.loc[df['X_COORD_CD'] == 0.0, ['X_COORD_CD', 'Y_COORD_CD']] = np.nan
df.loc[
    df['Latitude'].notna() & ((df['Latitude'] < 40.4) | (df['Latitude'] > 40.9)),
    ['X_COORD_CD', 'Y_COORD_CD', 'Latitude']
] = np.nan

print("STEP 2 DONE — Invalid coordinates set to NaN")
print("Remaining null Latitudes:", df['Latitude'].isnull().sum())
```

```python
date_cols = ['CMPLNT_FR_DT', 'CMPLNT_TO_DT', 'RPT_DT']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='mixed', errors='coerce')

# Fix time columns — extract only valid HH:MM:SS times
time_cols = ['CMPLNT_FR_TM', 'CMPLNT_TO_TM']
for col in time_cols:
    # Some time cells contain full datetime like "2026-01-07 12:30:00" — extract only time part
    extracted = df[col].astype(str).str.extract(r'(\d{2}:\d{2}:\d{2})', expand=False)
    df[col] = pd.to_datetime(extracted, format='%H:%M:%S', errors='coerce').dt.time

# Remove future/impossible dates (anything beyond 2025 is wrong for this dataset)
df.loc[df['CMPLNT_FR_DT'].dt.year > 2025, 'CMPLNT_FR_DT'] = np.nan
df.loc[df['RPT_DT'].dt.year > 2025, 'RPT_DT'] = np.nan

# Flag rows where complaint date is AFTER report date (temporal paradox)
df['IS_TEMPORAL_PARADOX'] = (
    df['CMPLNT_FR_DT'].notna() & 
    df['RPT_DT'].notna() & 
    (df['CMPLNT_FR_DT'] > df['RPT_DT'])
).astype(int)

print("STEP 3 DONE — Dates unified")
print("Temporal paradoxes flagged:", df['IS_TEMPORAL_PARADOX'].sum())
```

```python
valid_sex = ['M', 'F']
df['SUSP_SEX'] = df['SUSP_SEX'].str.strip().str.upper()
df['VIC_SEX'] = df['VIC_SEX'].str.strip().str.upper()
df.loc[~df['SUSP_SEX'].isin(valid_sex), 'SUSP_SEX'] = np.nan
df.loc[~df['VIC_SEX'].isin(valid_sex), 'VIC_SEX'] = np.nan

# Valid age groups
valid_age = ['<18', '18-24', '25-44', '45-64', '65+']
df['SUSP_AGE_GROUP'] = df['SUSP_AGE_GROUP'].str.strip().str.upper()
df['VIC_AGE_GROUP'] = df['VIC_AGE_GROUP'].str.strip().str.upper()
df.loc[~df['SUSP_AGE_GROUP'].isin(valid_age), 'SUSP_AGE_GROUP'] = np.nan
df.loc[~df['VIC_AGE_GROUP'].isin(valid_age), 'VIC_AGE_GROUP'] = np.nan

print("STEP 5 DONE — Entity codes removed from demographic fields")
print("Valid SUSP_SEX value counts:\n", df['SUSP_SEX'].value_counts())
```

```python
from sklearn.impute import KNNImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
#--- 6A: Impute BORO_NM using PATROL_BORO (mode within group) ----
df['BORO_NM'] = df.groupby('PATROL_BORO')['BORO_NM'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)
# Fill any still-remaining BORO_NM with overall mode
df['BORO_NM'] = df['BORO_NM'].fillna(df['BORO_NM'].mode()[0])

print("6A DONE — BORO_NM imputed")


# ---- 6B: Impute Coordinates using KNN ----
# Only impute if at least 2 of the 3 coordinate columns have data
knn_imputer = KNNImputer(n_neighbors=5)
df[['X_COORD_CD', 'Y_COORD_CD', 'Latitude']] = knn_imputer.fit_transform(
    df[['X_COORD_CD', 'Y_COORD_CD', 'Latitude']]
)

print("6B DONE — Coordinates imputed using KNN")


# ---- 6C: Impute Demographic Fields using Random Forest ----
# We use these columns as "clues" to predict missing demographics
feature_cols = ['LAW_CAT_CD', 'OFNS_DESC', 'PREM_TYP_DESC', 'BORO_NM', 'PATROL_BORO']

# Encode categorical features into numbers (Random Forest needs numbers)
encoders = {}
df_encoded = df.copy()
for col in feature_cols:
    le = LabelEncoder()
    # Fit only on non-null values, fill nulls with -1 temporarily
    df_encoded[col] = df_encoded[col].fillna('MISSING')
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    encoders[col] = le

def impute_with_rf(df, df_encoded, target_col, feature_cols):
    """Train RF on complete rows, predict missing rows."""
    
    # Rows where target is known
    train_mask = df[target_col].notna()
    # Rows where target is missing
    predict_mask = df[target_col].isna()
    
    if predict_mask.sum() == 0:
        print(f"  No missing values in {target_col}, skipping.")
        return df[target_col]
    
    if train_mask.sum() < 10:
        print(f"  Not enough data to train RF for {target_col}, skipping.")
        return df[target_col]
    
    X_train = df_encoded.loc[train_mask, feature_cols]
    y_train = df.loc[train_mask, target_col]
    X_predict = df_encoded.loc[predict_mask, feature_cols]
    
    rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
    rf.fit(X_train, y_train)
    
    predicted = rf.predict(X_predict)
    result = df[target_col].copy()
    result.loc[predict_mask] = predicted
    return result

# Apply RF imputation to demographic columns
for target in ['SUSP_AGE_GROUP', 'VIC_AGE_GROUP', 'SUSP_RACE', 'VIC_RACE', 'SUSP_SEX', 'VIC_SEX']:
    print(f"  Imputing {target}...")
    df[target] = impute_with_rf(df, df_encoded, target, feature_cols)

print("6C DONE — Demographics imputed using Random Forest")

# ---- 6D: Flag missing dates (don't impute, just mark) ----
df['IS_COMPLAINT_DATE_MISSING'] = df['CMPLNT_FR_DT'].isna().astype(int)

print("\n✅ ALL STEPS COMPLETE")
print("Final shape:", df.shape)
print("Remaining nulls:\n", df.isnull().sum())
```

```python
import pandas as pd
import numpy as np

# ============================================================
# PRE-STEP: Handle surviving nulls before feature engineering
# ============================================================

print("Null counts BEFORE final cleanup:")
print(df.isnull().sum())

# --- Categorical columns: fill remaining nulls with UNKNOWN ---
cat_cols = [
    'SUSP_RACE', 'VIC_RACE', 'SUSP_SEX', 'VIC_SEX',
    'SUSP_AGE_GROUP', 'VIC_AGE_GROUP', 'BORO_NM',
    'OFNS_DESC', 'PD_DESC', 'PREM_TYP_DESC',
    'LAW_CAT_CD', 'LOC_OF_OCCUR_DESC', 'PATROL_BORO',
    'CRM_ATPT_CPTD_CD', 'JURIS_DESC'
]
for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].fillna('UNKNOWN')

# --- Numeric columns: fill remaining nulls with median ---
num_cols = ['X_COORD_CD', 'Y_COORD_CD', 'Latitude', 
            'ADDR_PCT_CD', 'KY_CD', 'PD_CD']
for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].fillna(df[col].median())

# --- Drop columns where more than 20% is still null ---
threshold = 0.20
null_pct = df.isnull().sum() / len(df)
cols_to_drop = null_pct[null_pct > threshold].index.tolist()
# Keep date/time columns required for Feature 1 (Response Time in Hours)
cols_to_drop = [c for c in cols_to_drop if c not in ['CMPLNT_FR_DT', 'RPT_DT', 'CMPLNT_FR_TM']]
if cols_to_drop:
    print(f"\nDropping columns with >20% nulls: {cols_to_drop}")
    df = df.drop(columns=cols_to_drop)

print("\nNull counts AFTER final cleanup:")
print(df.isnull().sum())
print(f"\nDataset shape after cleanup: {df.shape}")
```

```python
# ============================================================
# FEATURE 1: Response Time in Hours
# How long between crime happening and being reported
# ============================================================

print("\n--- Engineering Feature 1: Response Time ---")

# Require date columns; they must not be dropped in PRE-STEP
if 'CMPLNT_FR_DT' not in df.columns or 'RPT_DT' not in df.columns:
    missing = [c for c in ['CMPLNT_FR_DT', 'RPT_DT'] if c not in df.columns]
    raise KeyError(
        f"Required date columns missing: {missing}. "
        "Re-run the PRE-STEP cell (so CMPLNT_FR_DT and RPT_DT are not dropped), "
        "or run the notebook from the top."
    )

# Make sure date columns are datetime type
df['CMPLNT_FR_DT'] = pd.to_datetime(df['CMPLNT_FR_DT'], errors='coerce')
df['RPT_DT'] = pd.to_datetime(df['RPT_DT'], errors='coerce')

# Calculate difference in hours
df['RESPONSE_TIME_HRS'] = (
    df['RPT_DT'] - df['CMPLNT_FR_DT']
).dt.total_seconds() / 3600

# Flag rows where response time is negative (temporal paradox)
df['IS_NEGATIVE_RESPONSE'] = (df['RESPONSE_TIME_HRS'] < 0).astype(int)
print(f"Temporal paradoxes found: {df['IS_NEGATIVE_RESPONSE'].sum()}")

# Set negative response times to null (they are errors)
df.loc[df['RESPONSE_TIME_HRS'] < 0, 'RESPONSE_TIME_HRS'] = np.nan

# Cap anything over 1 year (8760 hours) — clearly a data error
df.loc[df['RESPONSE_TIME_HRS'] > 8760, 'RESPONSE_TIME_HRS'] = np.nan

# Fill remaining nulls in response time with median
df['RESPONSE_TIME_HRS'] = df['RESPONSE_TIME_HRS'].fillna(
    df['RESPONSE_TIME_HRS'].median()
)

print("Response Time stats:")
print(df['RESPONSE_TIME_HRS'].describe().round(2))
```

```python
# ============================================================
# FEATURE 2: Time of Day Category
# Which part of the day did the crime happen
# ============================================================

print("\n--- Engineering Feature 2: Time of Day ---")

# Extract the hour from complaint time
# Handle cases where time might be a string like "14:30:00"
if df['CMPLNT_FR_TM'].dtype == object:
    df['COMPLAINT_HOUR'] = pd.to_datetime(
        df['CMPLNT_FR_TM'].astype(str),
        format='%H:%M:%S',
        errors='coerce'
    ).dt.hour
else:
    # If it's already a time object
    df['COMPLAINT_HOUR'] = pd.to_datetime(
        df['CMPLNT_FR_TM'].astype(str), 
        errors='coerce'
    ).dt.hour

# Fill missing hours with the most common hour
df['COMPLAINT_HOUR'] = df['COMPLAINT_HOUR'].fillna(
    df['COMPLAINT_HOUR'].mode()[0]
).astype(int)

# Bin hours into 4 time periods
df['TIME_OF_DAY'] = pd.cut(
    df['COMPLAINT_HOUR'],
    bins=[0, 6, 12, 18, 24],
    labels=['Night', 'Morning', 'Afternoon', 'Evening'],
    right=False,
    include_lowest=True
)

# Fill any remaining nulls in TIME_OF_DAY
df['TIME_OF_DAY'] = df['TIME_OF_DAY'].astype(str)
df['TIME_OF_DAY'] = df['TIME_OF_DAY'].replace('nan', 'Unknown')

print("Crime count by Time of Day:")
print(df['TIME_OF_DAY'].value_counts())

# Bonus: Felonies specifically by time of day
if 'LAW_CAT_CD' in df.columns:
    felony_mask = df['LAW_CAT_CD'].str.upper() == 'FELONY'
    print("\nFelonies by Time of Day:")
    print(df[felony_mask]['TIME_OF_DAY'].value_counts())
```

```python
# ============================================================
# FEATURE 3: Crime Severity Score
# Convert text category to a number: 1, 2, or 3
# ============================================================

print("\n--- Engineering Feature 3: Crime Severity Score ---")

severity_map = {
    'VIOLATION': 1,
    'MISDEMEANOR': 2,
    'FELONY': 3
}

# Map the values (convert to upper first to handle case issues)
df['LAW_CAT_CD'] = df['LAW_CAT_CD'].str.strip().str.upper()
df['CRIME_SEVERITY_SCORE'] = df['LAW_CAT_CD'].map(severity_map)

# For any that didn't match (UNKNOWN etc), fill with median score
median_severity = df['CRIME_SEVERITY_SCORE'].median()
df['CRIME_SEVERITY_SCORE'] = df['CRIME_SEVERITY_SCORE'].fillna(median_severity)

print("Severity score distribution:")
print(df['CRIME_SEVERITY_SCORE'].value_counts().sort_index())

# Average severity by borough
if 'BORO_NM' in df.columns:
    print("\nAverage Crime Severity by Borough:")
    print(
        df.groupby('BORO_NM')['CRIME_SEVERITY_SCORE']
        .mean().round(2)
        .sort_values(ascending=False)
    )
```

```python
# ============================================================
# BONUS FEATURE 4: Is Suspect Info Known?
# Binary flag — did we actually capture suspect details?
# ============================================================

print("\n--- Engineering Bonus Feature 4: Suspect Info Known ---")

df['SUSPECT_INFO_KNOWN'] = (
    (df['SUSP_AGE_GROUP'] != 'UNKNOWN') &
    (df['SUSP_RACE'] != 'UNKNOWN') &
    (df['SUSP_SEX'] != 'UNKNOWN')
).astype(int)

print(f"Rows with known suspect info: {df['SUSPECT_INFO_KNOWN'].sum()}")
print(f"Rows with unknown suspect info: {(df['SUSPECT_INFO_KNOWN'] == 0).sum()}")
```

```python
# ============================================================
# FINAL CHECK
# ============================================================

print("\n" + "="*50)
print("FEATURE ENGINEERING COMPLETE")
print("="*50)
print(f"Final dataset shape: {df.shape}")
print("\nNew columns added:")
new_cols = ['RESPONSE_TIME_HRS', 'IS_NEGATIVE_RESPONSE', 
            'COMPLAINT_HOUR', 'TIME_OF_DAY', 
            'CRIME_SEVERITY_SCORE', 'SUSPECT_INFO_KNOWN']
for col in new_cols:
    if col in df.columns:
        print(f"  ✅ {col}")

print("\nFinal null check:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# Save the final dataset
df.to_csv('final_sentinel_clean.csv', index=False)
print("\n✅ Final dataset saved as: final_sentinel_clean.csv")
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load your current cleaned file
df = pd.read_csv('final_sentinel_clean.csv', dtype=str, low_memory=False)
print(f"Starting shape: {df.shape}")


# ============================================================
# TASK 1: Handle Remaining Missing Data
# ============================================================

print("\n--- TASK 1: Missing Data ---")

# --- CMPLNT_FR_DT: fill missing from RPT_DT ---
df['CMPLNT_FR_DT'] = pd.to_datetime(df['CMPLNT_FR_DT'], errors='coerce')
df['RPT_DT'] = pd.to_datetime(df['RPT_DT'], errors='coerce')

missing_complaint_date = df['CMPLNT_FR_DT'].isna().sum()
print(f"Missing CMPLNT_FR_DT before fix: {missing_complaint_date}")

# Forward fill from RPT_DT where complaint date is missing
df.loc[df['CMPLNT_FR_DT'].isna(), 'CMPLNT_FR_DT'] = df.loc[
    df['CMPLNT_FR_DT'].isna(), 'RPT_DT'
]
print(f"Missing CMPLNT_FR_DT after fix: {df['CMPLNT_FR_DT'].isna().sum()}")

# --- CMPLNT_TO_DT: fill with CMPLNT_FR_DT where missing ---
df['CMPLNT_TO_DT'] = pd.to_datetime(df['CMPLNT_TO_DT'], errors='coerce')
df.loc[df['CMPLNT_TO_DT'].isna(), 'CMPLNT_TO_DT'] = df.loc[
    df['CMPLNT_TO_DT'].isna(), 'CMPLNT_FR_DT'
]

# --- CMPLNT_NUM: sequential imputation for missing IDs ---
# Find the max existing complaint number and assign new sequential IDs
df['CMPLNT_NUM'] = pd.to_numeric(df['CMPLNT_NUM'], errors='coerce')
missing_id_mask = df['CMPLNT_NUM'].isna()
missing_count = missing_id_mask.sum()
if missing_count > 0:
    max_id = df['CMPLNT_NUM'].max()
    new_ids = range(int(max_id) + 1, int(max_id) + 1 + missing_count)
    df.loc[missing_id_mask, 'CMPLNT_NUM'] = list(new_ids)
    print(f"Assigned {missing_count} new sequential IDs to missing CMPLNT_NUM")

# --- Drop rows only if they affect less than 5% of dataset ---
# Check which columns still have nulls
null_pct = df.isnull().sum() / len(df)
droppable_cols = null_pct[(null_pct > 0) & (null_pct < 0.05)].index.tolist()
print(f"\nColumns with <5% nulls (will drop those rows): {droppable_cols}")
before = len(df)
df = df.dropna(subset=droppable_cols)
after = len(df)
print(f"Rows dropped: {before - after} ({round((before-after)/before*100, 2)}%)")


# ============================================================
# TASK 2: Fix Data Types Properly
# ============================================================

print("\n--- TASK 2: Fix Data Types ---")

# --- Datetime columns ---
date_cols = ['CMPLNT_FR_DT', 'CMPLNT_TO_DT', 'RPT_DT']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
    print(f"  {col} → datetime ✅")

# --- Numeric columns ---
numeric_cols = ['ADDR_PCT_CD', 'KY_CD', 'PD_CD', 
                'X_COORD_CD', 'Y_COORD_CD', 'Latitude',
                'RESPONSE_TIME_HRS', 'COMPLAINT_HOUR', 
                'CRIME_SEVERITY_SCORE']
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        print(f"  {col} → numeric ✅")

# --- Cap PD_CD at realistic bounds ---
# Valid NYPD PD codes are between 0 and 999
# Anything above is a data entry error
if 'PD_CD' in df.columns:
    invalid_pdcd = df['PD_CD'] > 999
    print(f"\n  PD_CD values above 999 (invalid): {invalid_pdcd.sum()}")
    df.loc[invalid_pdcd, 'PD_CD'] = np.nan
    df['PD_CD'] = df['PD_CD'].fillna(df['PD_CD'].median())
    print(f"  PD_CD capped and nulls filled ✅")

# --- Cap ADDR_PCT_CD at realistic bounds ---
# NYC has precincts from 1 to 123
if 'ADDR_PCT_CD' in df.columns:
    invalid_pct = (df['ADDR_PCT_CD'] < 1) | (df['ADDR_PCT_CD'] > 123)
    print(f"  ADDR_PCT_CD out of bounds (1-123): {invalid_pct.sum()}")
    df.loc[invalid_pct, 'ADDR_PCT_CD'] = np.nan
    df['ADDR_PCT_CD'] = df['ADDR_PCT_CD'].fillna(df['ADDR_PCT_CD'].median())
    print(f"  ADDR_PCT_CD capped ✅")

# --- Categorical columns → category dtype ---
cat_cols = [
    'BORO_NM', 'LAW_CAT_CD', 'OFNS_DESC', 'PD_DESC',
    'PREM_TYP_DESC', 'PATROL_BORO', 'JURIS_DESC',
    'CRM_ATPT_CPTD_CD', 'LOC_OF_OCCUR_DESC',
    'SUSP_AGE_GROUP', 'SUSP_RACE', 'SUSP_SEX',
    'VIC_AGE_GROUP', 'VIC_RACE', 'VIC_SEX',
    'TIME_OF_DAY'
]
for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')
        print(f"  {col} → category dtype ✅")


# ============================================================
# TASK 3: Outlier & Validation Fixes
# ============================================================

print("\n--- TASK 3: Outliers & Validation ---")

# --- Filter pre-2025 and post-2025 dates ---
print("\n  Filtering invalid dates...")

for col in ['CMPLNT_FR_DT', 'RPT_DT']:
    before = len(df)
    # Keep only dates within 2024-2025 range (pilot dataset)
    invalid_dates = (
        df[col].dt.year < 2024
    ) | (
        df[col].dt.year > 2025
    )
    count = invalid_dates.sum()
    df.loc[invalid_dates, col] = np.nan
    print(f"  {col}: {count} dates outside 2024-2025 set to null")

# Re-fill nulled dates from RPT_DT
df.loc[df['CMPLNT_FR_DT'].isna(), 'CMPLNT_FR_DT'] = df.loc[
    df['CMPLNT_FR_DT'].isna(), 'RPT_DT'
]

# --- Validate lat/long within NYC bounds ---
print("\n  Validating coordinates within NYC bounds...")
if 'Latitude' in df.columns:
    # NYC bounds: lat 40.50 to 40.89
    invalid_lat = (
        df['Latitude'].notna()
    ) & (
        (df['Latitude'] < 40.50) | (df['Latitude'] > 40.89)
    )
    print(f"  Latitude out of NYC bounds: {invalid_lat.sum()}")
    df.loc[invalid_lat, 'Latitude'] = np.nan
    df.loc[invalid_lat, 'X_COORD_CD'] = np.nan
    df.loc[invalid_lat, 'Y_COORD_CD'] = np.nan

    # Fill remaining null coordinates with borough median
    df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
    if 'BORO_NM' in df.columns:
        df['Latitude'] = df.groupby('BORO_NM')['Latitude'].transform(
            lambda x: x.fillna(x.median())
        )
    print(f"  Coordinates cleaned ✅")

# --- Winsorize RESPONSE_TIME_HRS at 99th percentile ---
if 'RESPONSE_TIME_HRS' in df.columns:
    df['RESPONSE_TIME_HRS'] = pd.to_numeric(df['RESPONSE_TIME_HRS'], errors='coerce')
    p99 = df['RESPONSE_TIME_HRS'].quantile(0.99)
    p01 = df['RESPONSE_TIME_HRS'].quantile(0.01)
    before_max = df['RESPONSE_TIME_HRS'].max()
    df['RESPONSE_TIME_HRS'] = df['RESPONSE_TIME_HRS'].clip(lower=0, upper=p99)
    print(f"\n  RESPONSE_TIME_HRS winsorized:")
    print(f"    Before max: {before_max:.1f} hrs")
    print(f"    After max (99th pct): {p99:.1f} hrs")

    # Also create log-transformed version to reduce skewness
    df['RESPONSE_TIME_LOG'] = np.log1p(df['RESPONSE_TIME_HRS'])
    print(f"  RESPONSE_TIME_LOG created ✅")


# ============================================================
# TASK 4: Standardize Text in High-Cardinality Columns
# ============================================================

print("\n--- TASK 4: Text Standardization ---")

text_cols_to_clean = ['OFNS_DESC', 'PD_DESC', 'PREM_TYP_DESC']

for col in text_cols_to_clean:
    if col in df.columns:
        # Convert category back to string temporarily for cleaning
        df[col] = df[col].astype(str)

        # Lowercase, strip whitespace, remove double spaces
        df[col] = (
            df[col]
            .str.lower()
            .str.strip()
            .str.replace(r'\s+', ' ', regex=True)  # collapse multiple spaces
            .str.replace(r'[^\w\s,&/-]', '', regex=True)  # remove weird chars
        )

        unique_before = df[col].nunique()

        # Group rare categories (appear less than 10 times) into "other"
        value_counts = df[col].value_counts()
        rare_values = value_counts[value_counts < 10].index.tolist()
        df.loc[df[col].isin(rare_values), col] = 'other'

        unique_after = df[col].nunique()
        print(f"  {col}: {unique_before} → {unique_after} unique values after grouping rares ✅")

        # Convert back to category
        df[col] = df[col].astype('category')

# Standardize BORO_NM specifically (common source of inconsistency)
if 'BORO_NM' in df.columns:
    df['BORO_NM'] = df['BORO_NM'].astype(str).str.strip().str.upper()
    valid_boroughs = ['MANHATTAN', 'BROOKLYN', 'BRONX', 'QUEENS', 'STATEN ISLAND']
    df.loc[~df['BORO_NM'].isin(valid_boroughs), 'BORO_NM'] = 'UNKNOWN'
    df['BORO_NM'] = df['BORO_NM'].astype('category')
    print(f"\n  BORO_NM standardized to 5 valid boroughs ✅")
    print(f"  {df['BORO_NM'].value_counts().to_dict()}")


# ============================================================
# FINAL REPORT
# ============================================================

print("\n" + "="*55)
print("  FINAL DATA QUALITY REPORT")
print("="*55)

print(f"\nFinal shape: {df.shape}")

print("\nData types:")
print(df.dtypes.value_counts())

null_remaining = df.isnull().sum()
null_remaining = null_remaining[null_remaining > 0]
if len(null_remaining) == 0:
    print("\n✅ Zero nulls remaining in dataset!")
else:
    print(f"\nRemaining nulls (if any):")
    print(null_remaining)

print("\nColumn-by-column summary:")
for col in df.columns:
    null_count = df[col].isnull().sum()
    null_pct = round(null_count / len(df) * 100, 2)
    dtype = df[col].dtype
    print(f"  {col:<35} dtype={str(dtype):<12} nulls={null_count} ({null_pct}%)")

# Save the truly final file
df.to_csv('final_sentinel_v2.csv', index=False)
print("\n✅ Saved as: final_sentinel_v2.csv")
print("\nThis file is approximately 8.5/10 clean.")
```

```python
def chat(client, system_prompt, conversation_history, user_message):
    
def load_data(path):
    """Load the CSV and return a DataFrame."""
    try:
        df = pd.read_csv(path, dtype=str, low_memory=False)
        print(f"✅ Loaded dataset: {df.shape[0]} rows, {df.shape[1]} columns\n")
        return df
    except FileNotFoundError:
        print(f"❌ File not found: {path}")
        print("   Make sure your cleaned CSV is in the same folder as this script.")
        exit()


def compute_stats(df):
    """
    Pre-compute key statistics from the dataframe.
    This is what we send to Claude so it can answer questions
    without needing to see every single row.
    """
    stats = {}

    # --- Basic shape ---
    stats['total_records'] = len(df)
    stats['columns'] = list(df.columns)

    # --- Borough breakdown ---
    if 'BORO_NM' in df.columns:
        stats['crimes_by_borough'] = (
            df['BORO_NM'].str.strip().str.upper()
            .value_counts().to_dict()
        )

    # --- Crime category breakdown ---
    if 'LAW_CAT_CD' in df.columns:
        stats['crimes_by_category'] = (
            df['LAW_CAT_CD'].str.strip().str.upper()
            .value_counts().to_dict()
        )

    # --- Top 10 offense types ---
    if 'OFNS_DESC' in df.columns:
        stats['top_10_offenses'] = (
            df['OFNS_DESC'].str.strip().str.upper()
            .value_counts().head(10).to_dict()
        )

    # --- Suspect demographics ---
    if 'SUSP_AGE_GROUP' in df.columns:
        stats['suspect_age_distribution'] = (
            df['SUSP_AGE_GROUP'].str.strip().str.upper()
            .value_counts().to_dict()
        )

    if 'SUSP_RACE' in df.columns:
        stats['suspect_race_distribution'] = (
            df['SUSP_RACE'].str.strip().str.upper()
            .value_counts().to_dict()
        )

    if 'SUSP_SEX' in df.columns:
        stats['suspect_sex_distribution'] = (
            df['SUSP_SEX'].str.strip().str.upper()
            .value_counts().to_dict()
        )

    # --- Victim demographics ---
    if 'VIC_AGE_GROUP' in df.columns:
        stats['victim_age_distribution'] = (
            df['VIC_AGE_GROUP'].str.strip().str.upper()
            .value_counts().to_dict()
        )

    if 'VIC_RACE' in df.columns:
        stats['victim_race_distribution'] = (
            df['VIC_RACE'].str.strip().str.upper()
            .value_counts().to_dict()
        )

    if 'VIC_SEX' in df.columns:
        stats['victim_sex_distribution'] = (
            df['VIC_SEX'].str.strip().str.upper()
            .value_counts().to_dict()
        )

    # --- Time of day breakdown (if feature engineered) ---
    if 'TIME_OF_DAY' in df.columns:
        stats['crimes_by_time_of_day'] = (
            df['TIME_OF_DAY'].value_counts().to_dict()
        )

    # --- Felonies by time of day ---
    if 'TIME_OF_DAY' in df.columns and 'LAW_CAT_CD' in df.columns:
        felony_mask = df['LAW_CAT_CD'].str.strip().str.upper() == 'FELONY'
        stats['felonies_by_time_of_day'] = (
            df[felony_mask]['TIME_OF_DAY'].value_counts().to_dict()
        )

    # --- Crimes by premise type ---
    if 'PREM_TYP_DESC' in df.columns:
        stats['top_10_premises'] = (
            df['PREM_TYP_DESC'].str.strip().str.upper()
            .value_counts().head(10).to_dict()
        )

    # --- Crimes by patrol borough ---
    if 'PATROL_BORO' in df.columns:
        stats['crimes_by_patrol_boro'] = (
            df['PATROL_BORO'].str.strip().str.upper()
            .value_counts().to_dict()
        )

    # --- Response time stats (if feature engineered) ---
    if 'RESPONSE_TIME_HRS' in df.columns:
        df['RESPONSE_TIME_HRS'] = pd.to_numeric(df['RESPONSE_TIME_HRS'], errors='coerce')
        valid_rt = df['RESPONSE_TIME_HRS'].dropna()
        valid_rt = valid_rt[valid_rt >= 0]
        if len(valid_rt) > 0:
            stats['response_time_stats'] = {
                'mean_hours': round(valid_rt.mean(), 2),
                'median_hours': round(valid_rt.median(), 2),
                'max_hours': round(valid_rt.max(), 2)
            }

    # --- Crime severity score (if feature engineered) ---
    if 'CRIME_SEVERITY_SCORE' in df.columns:
        df['CRIME_SEVERITY_SCORE'] = pd.to_numeric(df['CRIME_SEVERITY_SCORE'], errors='coerce')
        if 'BORO_NM' in df.columns:
            severity_by_boro = (
                df.groupby('BORO_NM')['CRIME_SEVERITY_SCORE']
                .mean().round(2).dropna().to_dict()
            )
            stats['avg_severity_by_borough'] = severity_by_boro

    # --- Null counts (data quality info) ---
    stats['null_counts'] = df.isnull().sum().to_dict()

    return stats


def build_system_prompt(stats):
    """
    Build the system prompt that tells Claude what dataset it is working with
    and gives it all the pre-computed statistics.
    """
    prompt = f"""
You are a data analyst assistant for the NYPD Sentinel Pilot 2025 crime dataset.
You have been given pre-computed statistics from a cleaned crime complaints CSV file.
Use ONLY the statistics provided below to answer user questions.
Be specific, cite numbers, and give clear insights.
If a question cannot be answered from the available stats, say so honestly.

=== DATASET OVERVIEW ===
Total Records: {stats.get('total_records', 'N/A')}
Columns available: {', '.join(stats.get('columns', []))}

=== CRIMES BY BOROUGH ===
{stats.get('crimes_by_borough', 'Not available')}

=== CRIMES BY LEGAL CATEGORY (Felony / Misdemeanor / Violation) ===
{stats.get('crimes_by_category', 'Not available')}

=== TOP 10 OFFENSE TYPES ===
{stats.get('top_10_offenses', 'Not available')}

=== CRIMES BY TIME OF DAY ===
{stats.get('crimes_by_time_of_day', 'Not available')}

=== FELONIES BY TIME OF DAY ===
{stats.get('felonies_by_time_of_day', 'Not available')}

=== TOP 10 PREMISES WHERE CRIMES OCCUR ===
{stats.get('top_10_premises', 'Not available')}

=== CRIMES BY PATROL BOROUGH ===
{stats.get('crimes_by_patrol_boro', 'Not available')}

=== SUSPECT AGE DISTRIBUTION ===
{stats.get('suspect_age_distribution', 'Not available')}

=== SUSPECT RACE DISTRIBUTION ===
{stats.get('suspect_race_distribution', 'Not available')}

=== SUSPECT SEX DISTRIBUTION ===
{stats.get('suspect_sex_distribution', 'Not available')}

=== VICTIM AGE DISTRIBUTION ===
{stats.get('victim_age_distribution', 'Not available')}

=== VICTIM RACE DISTRIBUTION ===
{stats.get('victim_race_distribution', 'Not available')}

=== VICTIM SEX DISTRIBUTION ===
{stats.get('victim_sex_distribution', 'Not available')}

=== RESPONSE TIME STATISTICS ===
{stats.get('response_time_stats', 'Not available (feature may not have been engineered)')}

=== AVERAGE CRIME SEVERITY SCORE BY BOROUGH ===
(Scale: Violation=1, Misdemeanor=2, Felony=3)
{stats.get('avg_severity_by_borough', 'Not available (feature may not have been engineered)')}

=== NULL / MISSING VALUE COUNTS PER COLUMN ===
{stats.get('null_counts', 'Not available')}
"""
    return prompt.strip()


def chat(client, system_prompt, conversation_history, user_message):
    """
    Send the user's message along with full conversation history to Claude.
    Returns Claude's reply as a string.
    """
    # Add the new user message to history
    conversation_history.append({
        "role": "user",
        "content": user_message
    })

    response = client.messages.create(
        model="claude-sonnet-4-20250514",
        max_tokens=1024,
        system=system_prompt,
        messages=conversation_history
    )

    assistant_reply = response.content[0].text

    # Add Claude's reply to history so next question has context
    conversation_history.append({
        "role": "assistant",
        "content": assistant_reply
    })

    return assistant_reply


def main():
    print("=" * 60)
    print("   NYPD SENTINEL 2025 — Crime Data Chatbot")
    print("   Powered by Claude API")
    print("=" * 60)
    print()

    # 1. Load data
    df = load_data(CSV_PATH)

    # 2. Compute statistics from the dataframe
    print("⏳ Computing statistics from dataset...")
    stats = compute_stats(df)
    print("✅ Statistics ready.\n")

    # 3. Build system prompt with all stats embedded
    system_prompt = build_system_prompt(stats)

    # 4. Initialize Claude client
    client = anthropic.Anthropic(api_key=API_KEY)

    # 5. Conversation history — grows as you chat
    conversation_history = []

    print("💬 Chatbot is ready! Type your question below.")
    print("   Type 'exit' or 'quit' to stop.\n")
    print("-" * 60)

    # 6. Chat loop
    while True:
        user_input = input("\nYou: ").strip()

        if not user_input:
            continue

        if user_input.lower() in ['exit', 'quit', 'bye']:
            print("\nGoodbye! 👋")
            break

        print("\nClaude: ", end="", flush=True)
        reply = chat(client, system_prompt, conversation_history, user_input)
        print(reply)
        print("-" * 60)


if __name__ == "__main__":
    main()
    conversation_history.append({
        "role": "user",
        "content": user_message
    })

    response = client.chat.completions.create(
        model="perplexity/sonar",
        messages=[
            {"role": "system", "content": system_prompt},
            *conversation_history
        ]
    )

    assistant_reply = response.choices[0].message.content

    conversation_history.append({
        "role": "assistant",
        "content": assistant_reply
    })

    return assistant_reply
```
